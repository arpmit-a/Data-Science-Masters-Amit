```python
#Q3
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Range1d

# Enable Bokeh output in the notebook
output_notebook()

# Prepare the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title="Scatter Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add circle glyphs to the figure with customizations
p.circle(x, y, size=10, color="blue", alpha=0.5)

# Set the title of the plot
p.title.text = "My Plot"

# Customize the axis labels
p.xaxis.axis_label = "Custom X-axis Label"
p.yaxis.axis_label = "Custom Y-axis Label"

# Modify the appearance of the axis ticks
p.xaxis.major_label_text_color = "red"
p.yaxis.major_label_text_font_size = "12pt"

# Adjust the range or scaling of the axes
p.x_range = Range1d(0, 10)  # Set the x-axis range
p.y_range.start = -5  # Set the starting value of the y-axis

# Add a legend to the plot
p.legend.location = "top_left"
p.legend.title = "Categories"
p.legend.label_text_font_size = "10pt"

# Customize the color and style of the glyphs
p.circle(x, y, size=10, color="red", alpha=0.8)
p.line(x, y, line_width=2, line_dash="dashed", line_color="blue")

# Display the plot
show(p)
```

```python
#Q4
'''The Bokeh server is a key component of the Bokeh library that allows you to create and deploy interactive plots and applications that can be updated in real time. It enables you to build web applications with interactive visualizations using Python.

When you run a Bokeh server, it acts as a communication bridge between the Python code running on the server and the web browser running on the client side. 
The server maintains a WebSocket connection with the client, allowing for bi-directional communication.

Using the Bokeh server, you can create interactive plots that respond to user interactions such as mouse clicks, mouse movements, and widget interactions.
When an event is triggered, the server-side Python code can update the data or properties of the plot, and those changes are automatically sent to the client-side web browser, resulting in real-time updates to the plot.'''

'''To use the Bokeh server to create interactive plots, you typically follow these steps:

Import the necessary modules from Bokeh.
Define the plot and its initial configuration.
Define the update functions that modify the plot data or properties based on user interactions or other events.
Create the Bokeh application by associating the update functions with the plot.
Run the Bokeh server, which starts a web server and makes the application available in a web browser.'''

#Here's an example how to do it
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.events import DoubleTap

# Create the figure and data source
p = figure()
source = ColumnDataSource(data=dict(x=[], y=[]))

# Define the update function
def update_plot():
    # Update the data or properties of the plot
    # Here, we generate random data for demonstration
    new_data = dict(x=[1, 2, 3], y=[4, 5, 6])
    source.data = new_data

# Associate the update function with the plot
p.on_event(DoubleTap, lambda: update_plot())

# Add the plot to the document
curdoc().add_root(p)

# Run the Bokeh server
# Open the app in a browser by visiting http://localhost:5006/ in most cases
```
