```python
#q1
Estimation statistics refers to methods that use sample data to estimate characteristics or parameters of a population. 
The two main types of estimates are:
Point estimate: A single numerical value that provides an estimate of a population parameter. For example, using the sample mean as an estimate of the population mean.
Interval estimate: An estimate in the form of an interval that is likely to contain the true parameter value. For example, a 95% confidence interval gives a range of likely values for the population parameter.
```

```python
#q2
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    # Calculate standard error of the mean
    standard_error = sample_std / math.sqrt(sample_size)
    
    # Calculate margin of error (assuming a 95% confidence interval)
    margin_of_error = 1.96 * standard_error
    
    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (lower_bound, upper_bound)

# Example usage:
sample_mean = 75
sample_std = 10
sample_size = 30

confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size)
print("Estimated population mean (95% confidence interval):", confidence_interval)
```

```python
#q3
Hypothesis testing is a statistical method that uses sample data to evaluate assumptions or claims about a population parameter. It allows us to make statistical decisions using experimental data.

Importance of hypothesis testing:

It helps determine if results provide enough evidence to reject the null hypothesis or support the alternative hypothesis
1.Enables making data-driven decisions in science, business, and social research
2.Tests assumptions, theories, and results to add statistical rigor and evidence
3.Useful for statistical inference to make claims about populations
```

```python
#q4
Null hypothesis (H0): The average weight of male college students is less than or equal to the average weight of female college students.

Alternative hypothesis (Ha): The average weight of male college students is greater than the average weight of female college students.
```

```python
#q5
import scipy.stats as stats

# Assuming you have two sample datasets: data_male and data_female
# Example datasets:
data_male = [70, 72, 75, 73, 71, 74, 76, 78, 80]
data_female = [65, 68, 72, 69, 67, 70, 73, 75, 77]

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(data_male, data_female)

# Set the significance level (alpha)
alpha = 0.05

# Compare p-value to alpha
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence that the average weight of male college students is greater than the average weight of female college students.")
else:
    print("Fail to reject the null hypothesis. The data does not provide sufficient evidence to conclude that the average weight of male college students is greater than the average weight of female college students.")
```

```python
#q6
A null hypothesis (H0) is a statement of no statistical significance or no effect. It states that there is no difference between a sample statistic and a population parameter. The alternative hypothesis (Ha) is a claim that there is a statistical significance or effect in the population.

Some examples:

H0: There is no difference between the mean heights of men and women.

Ha: The mean height of men is greater than that of women.

H0: A new medicine does not improve memory over placebo.
Ha: The new medicine improves memory compared to placebo.
```

```python
#q7
The main steps in hypothesis testing are:

Make assumptions and state the null and alternative hypotheses
Set a significance level (common levels are 5% or 1%)
Calculate the test statistic (z-score, t-score, F-ratio)
Determine the p-value
Compare the p-value against the significance level and make a decision (reject or fail to reject null hypothesis)
Interpret results in context of the problem
```

```python
#q8
The p-value is the probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis is true. 
A smaller p-value provides stronger evidence against the null hypothesis, indicating it should be rejected. 
A p-value less than the significance level set (commonly 0.05) means the observed result would be highly unlikely under the null hypothesis, so we reject the null.
The p-value helps determine statistical significance in hypothesis testing.
```

```python
#q10
import scipy.stats as stats
import numpy as np

# Sample data  
sample1 = np.random.normal(size=30) 
sample2 = np.random.normal(size=30)

# Null hypothesis  
H0 = "The population means are equal" 

# Calculate test statistic
t_stat, p_val = stats.ttest_ind(sample1, sample2)  

# Interpret p-value
alpha = 0.05
if p_val < alpha:
  print("Reject H0")  
else:
  print("Fail to reject H0") 

print("p-value: ", p_val)
```

```python
#q11
Student's t-distribution is used for hypothesis testing when the population standard deviation is unknown and sample sizes are small. 
It provides a more conservative analysis compared to the normal distribution. 
The t-distribution has wider tails, giving more probability to extreme values. 
As degrees of freedom increase, the t-distribution approaches the normal distribution.
Student's t-tests are commonly used to compare sample means.
```

```python
#q12
The t-statistic is a ratio of the difference between the sample mean and null hypothesis mean to the estimated standard error. It represents the number of standard deviations the sample mean is from the hypothesized value in a t-test. Higher t-statistic absolute values indicate further deviation.

Formula:
t = (Sample mean - Null mean) / (Standard error)
```

```python
#q13
Sample mean (x) = $500

Sample standard deviation (s) = $50
Sample size (n) = 50

To calculate a 95% confidence interval for the mean daily revenue :

Standard error = s / n = $50 / 50 = $7.07

Margin of error = z*  SE = 1.96  $7.07 = $13.87

95% confidence interval = x  margin of error
= $500  $13.87
= ($486.13, $513.87)

We estimate with 95% confidence that the true average daily revenue lies between $486.13 and $513.87.
```
