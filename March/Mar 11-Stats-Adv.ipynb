```python
#Q1: Difference between t-test and z-test:
The key difference between a t-test and a z-test lies in the type of data being analyzed and the population standard deviation.

Scenario for t-test:

Example: Suppose you want to compare the mean scores of two groups (e.g., Group A and Group B) in a small sample size (typically less than 30). You would use a t-test if the population standard deviation is unknown or if the sample size is small. For instance, comparing the average test scores of two different teaching methods in a class of 20 students.
Scenario for z-test:

Example: If you have a large sample size (usually more than 30) and you know the population standard deviation, you might use a z-test. For instance, assessing whether the average height of a sample of 1000 people is significantly different from the known population average height.
```

```python
# Q2: One-tailed vs. two-tailed tests:

In a one-tailed test, you're interested in only one direction of difference or effect (either greater than or less than).

Example: Testing whether a new drug improves performance, with the alternative hypothesis stating it will increase performance.
In a two-tailed test, you're interested in differences or effects in both directions (greater than or less than).

Example: Examining whether a new teaching method has any effect on test scores, without specifying whether the effect will be an increase or decrease.
```

```python
#Q3: Type 1 and Type 2 errors:

Type 1 error (False Positive): Rejecting a true null hypothesis.

Example: Concluding a new drug is effective when it has no actual effect.
Type 2 error (False Negative): Failing to reject a false null hypothesis.

Example: Concluding a new drug has no effect when it is genuinely effective.
```

```python
# Q4. 
Bayes' theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event. For example, if cancer is related to age, Bayes' theorem allows us to determine the probability that a person has cancer, given their age.

The theorem is stated as:
P(A|B) = P(B|A) * P(A) / P(B)
Where:
P(A|B) is the probability of A given that B is true.
P(B|A) is the probability of B given that A is true.
P(A) is the probability of A being true (prior probability).
P(B) is the probability of B being true.

So if A is having cancer and B is being over 50 years old, Bayes' theorem allows us to calculate the probability of having cancer given someone is over 50 years old.
```

```python
Q5. A confidence interval gives a range of values within which we can say, with a certain confidence level, that the true population parameter lies. To calculate a 95% confidence interval, we use the sample mean, sample standard deviation, and the z-score associated with the desired confidence level. 

For example, if we collect a random sample with n=100, a sample mean of x̄ = 50, and a sample standard deviation of s = 5, the 95% confidence interval is:

(x̄ - z∗ × SE, x̄ + z∗ × SE)

Where z* = 1.96 for a 95% CI, and the standard error SE = s / √n = 5 / 10 = 0.5

So the 95% CI is (50 - 1.96*0.5, 50 + 1.96*0.5) = (48.98, 51.02)

We are 95% confident the true population mean lies between 48.98 and 51.02.
```

```python
#Q6
Here is an example of using Bayes' theorem to calculate a probability with prior knowledge and new evidence:

Suppose there is a test to detect a rare disease. The probability someone has the disease is 0.1% (prior probability of A). The test has a 98% chance of correctly detecting the disease if someone has it (probability of B given A). However, it also yields a positive result in 5% of cases where the person does not have the disease (probability of B given not A).

We want to know, if a person tests positive, what is the probability they actually have the disease (posterior probability of A given B)?

P(A) = 0.001  (0.1% have the disease)
P(B|A) = 0.98 (98% test sensitivity) 
P(B|not A) = 0.05 (5% false positive rate)
P(A|B) = probability we want to calculate

Using Bayes' Theorem:
P(A|B) = P(B|A) * P(A) / P(B)

To calculate P(B):
P(B) = P(B|A)*P(A) + P(B|not A)*P(not A) 
= 0.98 * 0.001 + 0.05 * 0.999
= 0.00098 + 0.04995
= 0.05093

Plugging into Bayes' Theorem: 
P(A|B) = P(B|A) * P(A) / P(B)
= 0.98 * 0.001 / 0.05093
= 0.019

Therefore, the probability of having the disease given a positive test is 1.9%, even though the test is quite accurate. This demonstrates the effect of prior probability on the interpretation of new evidence.
```

```python
Q7. The 95% confidence interval for a sample with mean 50 and standard deviation 5, with sample size n, is calculated as:

CI = (mean - margin of error, mean + margin of error)

The margin of error (ME) is calculated as: 
ME = z-score * (std dev / √n)

For a 95% CI, z = 1.96.
Std dev = 5 
n = not provided, so assuming n = 36

ME = 1.96 * (5 / √36) = 1.96 * (5 / 6) = 1.63

Therefore, the 95% CI is:
(50 - 1.63, 50 + 1.63) = (48.37, 51.63)

We can say with 95% confidence that the true population mean lies between 48.37 and 51.63 based on this sample.
```

```python
Q8. The margin of error (ME) indicates the maximum amount that the sample mean will differ from the true population mean. It depends directly on the sample standard deviation and inversely on the square root of the sample size. As the sample size increases, the ME will decrease.

For example, if estimating the average height in a population with a std dev of 3 inches, using a sample size of 25 would give an ME of: 

ME = 1.96 * (3 / √25) = 1.96 * (3 / 5) = 1.18 inches

But with a sample of 100, the ME is:

ME = 1.96 * (3 / √100 ) = 1.96 * (3 / 10) = 0.59 inches

So a larger sample size of 100 versus 25 reduces the margin of error from 1.18 inches to 0.59 inches. The estimate will be more precise with the larger sample.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

To calculate the z-score, we use the formula:

z = (X - μ) / σ
Where:
X = 75 (value of the data point)  
μ = 70 (population mean)
σ = 5 (population standard deviation)

Plugging this into the formula:
z = (75 - 70) / 5 = 1

A z-score of 1 indicates that the data point is 1 standard deviation above the mean. So a value of 75 is higher than the average value of 70 in this population.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

Ho: μ = 0  (The drug is not effective, i.e. the average weight loss is 0)  
Ha: μ > 0 (The drug is effective, i.e. the average weight loss is greater than 0)

Test statistic: 
t = (sample mean - hypothesized mean) / (standard error)
= (6 - 0) / (2.5 / √50) = 12

Degrees of freedom = n - 1 = 50 - 1 = 49

Using a t-table, the critical value at df = 49 and 95% confidence level is 1.68.

Since the test statistic t = 12 > critical value 1.68, we reject the null hypothesis. 

There is sufficient evidence at the 5% significance level to conclude that the new weight loss drug is effective.

```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

Sample proportion (p') = 0.65

Standard error = √(p'(1-p')/n) 
           = √(0.65(1-0.65)/500)
           = 0.02

Margin of error = 1.96 * standard error 
           = 1.96 * 0.02  
           = 0.039
           
Confidence interval = p' ± margin of error
                  = 0.65 ± 0.039  
                  = (0.611, 0.689)
                  
The 95% confidence interval for the true proportion who are satisfied is 0.611 to 0.689.                  

```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

Sample mean (X̄) = 65  
Sample standard deviation (s) is not provided, use population standard deviation (σ) = 8  
Sample size (n) = 50

Margin of error = z-score * σ/√n
             = 1.645 * 8/√50 (z-score for 90% confidence level)
             = 1.645 * 1.13
             = 1.86
             
Confidence interval = X̄ ± margin of error 
                   = 65 ± 1.86
                   = (63.14, 66.86)
                   
The 90% confidence interval for the true population mean is 63.14 to 66.86.

```

```python
Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

Null hypothesis (H0): Caffeine does not affect reaction time (μ = 0)
Alternative hypothesis (Ha): Caffeine affects reaction time (μ ≠ 0)  

Test statistic: 
t = (sample mean - hypothesized mean) / (standard error)

Sample mean (X̄) = 0.25
Standard deviation (s) = 0.05  
Sample size (n) = 30

Standard error = s / √n 
           = 0.05 / √30
           = 0.009
           
Plugging into test statistic:
t = (0.25 - 0) / 0.009 = 27.8

The critical value at 90% confidence level with df = n - 1 = 29 is 1.70.

Since t = 27.8 > 1.70, we reject the null hypothesis.

There is sufficient evidence to conclude the caffeine has a significant effect on reaction time at the 10% significance level.
```
