```python
#Q1
 Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides an extensive collection of tools for creating various types of plots, charts, and graphs.

Five plots that can be created using the Pyplot module of Matplotlib are:

Line plot: Used to display data points connected by lines.
Scatter plot: Used to display data points as individual dots, with each dot representing a data point.
Bar plot: Used to display data using rectangular bars, with the height or length of the bar representing the data value.
Histogram: Used to display the distribution of a dataset by dividing it into bins and displaying the frequency of data points in each bin.
Pie chart: Used to display data in a circular chart, with each slice representing a data category and the size of the slice representing the data value.
```

```python
#Q2
# One variable is plotted on the x-axis and another variable is plotted on the y-axis in a scatter plot, which depicts two-dimensional data points as individual dots on a graph. 
# It helps to spot any patterns or trends by visualising the relationship between two variables.
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)  # Create a scatter plot
plt.title("Scatter Plot")  # Add title
plt.xlabel("X Axis Label")  # Add x-axis label
plt.ylabel("Y Axis Label")  # Add y-axis label
plt.show()  # Display the plot
```

```python
#Q3
'''To include many plots in a single figure, use Matplotlib's subplot() function. 
It enables the simultaneous comparison and analysis of numerous plots by grouping them in a grid-like style with rows and columns.'''

import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 subplot layout
fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot line 1
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Line1')

# Plot line 2
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Line2')

# Plot line 3
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Line3')

# Plot line 4
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Line4')

# Add overall title
fig.suptitle('Multiple Line Plots')

# Display the plots
plt.show()

```

```python
#Q4
'''A bar plot is a type of graph that uses rectangular bars to depict data, with the height or length of the bars serving as the values of the data. 
It is frequently employed to contrast and present categorical data, such as when contrasting various categories or groupings.'''
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title('Bar Plot: Company vs Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title('Horizontal Bar Plot: Company vs Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

```python
#Q5
'''A box plot, also known as a whisker plot, is a graphical representation of a dataset's summary statistics, including the median, quartiles, and outliers. 
It is used to display the distribution and variability of the data, identify skewness or asymmetry, and detect potential outliers.'''
import matplotlib.pyplot as plt
import numpy as np

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

# Create a box plot
ax.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
ax.set_title('Box Plot')
ax.set_xlabel('Boxes')
ax.set_ylabel('Values')

# Display the plot
plt.show()
```
